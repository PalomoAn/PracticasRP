```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar los datos
url = "http://robjhyndman.com/tsdldata/data/nybirths.dat"
nacimientos = pd.read_table(url, header=None)
```

```python

# Crear los datos de la serie temporal
nacimientos.index = pd.date_range(start="1946-01", periods=len(nacimientos), freq='M')
nacimientos.columns = ['Nacimientos']

# Crear una columna 'Mes' para agrupar por mes
nacimientos['Mes'] = nacimientos.index.month
```

```python
# Crear el boxplot
plt.figure(figsize=(10, 6))
nacimientos.boxplot(column='Nacimientos', by='Mes', vert=False)
plt.title("Boxplot de Nacimientos por Mes")
plt.suptitle("")  # Eliminar el título por defecto
plt.xlabel("Número de Nacimientos")
plt.ylabel("Mes")
plt.grid(True)
plt.show()
```
