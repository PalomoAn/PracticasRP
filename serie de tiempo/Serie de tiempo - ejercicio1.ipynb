```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.tseries.offsets import MonthBegin
```

```python
# Cargar los datos
url = "http://robjhyndman.com/tsdldata/data/nybirths.dat"
nacimientos = pd.read_table(url, header=None)

# Crear los datos de la serie temporal
nacimientos.index = pd.date_range(start="1946-01", periods=len(nacimientos), freq='M')
nacimientos.columns = ['Nacimientos']
```

```python
# Graficar la serie temporal
plt.figure(figsize=(10, 5))
plt.plot(nacimientos.index, nacimientos['Nacimientos'], label="Nacimientos en NY")
plt.title("Nacimientos mensuales en Nueva York (1946-1959)")
plt.xlabel("Año")
plt.ylabel("Número de nacimientos")
plt.legend()
plt.grid()
plt.show()
```
